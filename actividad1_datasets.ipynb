```python
import pandas as pd
ruta='/home/angie/Escritorio/repos/BITrepository/'
tabla=pd.read_csv(ruta+'DATA/Desaparecidos._Colombia__hist_rico__a_os_1930_a_marzo_de_2024_20240729.csv',sep=',',encoding='utf-8')
tabla
```

```python
tabla.info()
```

```python
# Hacer el conteo por genero y estado de la desaparicion
conteo = tabla.groupby(['Sexo del desaparecido', 'Estado de la desaparición']).size().reset_index(name='Cantidad')
print(conteo)
```

```python
# Hacer el conteo por genero y nacionalidad
conteo = tabla.groupby(['Sexo del desaparecido', 'País de nacimiento del desaparecido']).size().reset_index(name='Cantidad')
print(conteo)
```

```python
import matplotlib.pyplot as plt

conteo = conteo.sort_values(by='Cantidad', ascending=False)

# Crear el gráfico de torta
plt.figure(figsize=(8, 6))
conteo.plot(kind='pie', y='Cantidad', labels=conteo['Sexo del desaparecido'] + ' - ' + conteo['País de nacimiento del desaparecido'], autopct='%1.1f%%', startangle=100)

# Agregar título y mostrar el gráfico
plt.title('Clasificación de la desaparición por sexo y país')
plt.axis('equal')  # Asegurar que el gráfico sea circular
plt.show()
```

**VARIABLES DE INTERES**

 **Fecha de la desaparición** Necesarias para identificar los años con mayor/menor desaparición 
 **Sexo del desaparecido** Identificar sexo con mas/menos desapariciones 
 **País de nacimiento del desaparecido** De que paises son las personas desparecidas
 **Departamento donde ocurre la desaparición DANE** Departamento con mayor/menor desapariciones
