```python
import pandas as pd
import os 

path=os.getcwd()
print(path)
file="Desaparecidos._Colombia__hist_rico__a_os_1930_a_marzo_de_2024_20240729.csv"
full_path = os.path.join(path,'DATA', file)
print(full_path)
tabla=pd.read_csv(full_path,sep=',',encoding='utf-8')
tabla.head()
```

**1. Ajustar nombre de columnas**

```python
tabla.rename(columns={"Departamento donde ocurre la desaparición DANE": "Departamento"}, inplace=True)
tabla.rename(columns={"Estado de la desaparición":"Estado"}, inplace=True)
tabla.rename(columns={"Clasificación de la desaparición":"Clasificacion"}, inplace=True)
tabla.rename(columns={"Fecha de la desaparición":"Fecha_desaparición"}, inplace=True)
tabla.rename(columns={"Edad al momento de la desaparición":"Edad"}, inplace=True)
tabla.rename(columns={"Sexo del desaparecido":"Sexo"}, inplace=True)
tabla.rename(columns={"País de nacimiento del desaparecido":"Pais_nacimiento"}, inplace=True)
tabla.rename(columns={"Ancestro racial del desaparecido":"Raza"}, inplace=True)
tabla.rename(columns={"Estado civil del desaparecido":"Estado_civil"}, inplace=True)
tabla.rename(columns={"Escolaridad del desaparecido":"Escolaridad"}, inplace=True)
tabla.rename(columns={"Pertenencia grupal del desaparecido":"Pertenencia_grupal"}, inplace=True)
tabla.rename(columns={"País donde ocurre la desaparición":"Pais_desaparicion"}, inplace=True)
tabla.rename(columns={"Municipio donde ocurre la desaparición DANE":"Municipio"}, inplace=True)

tabla.info()
```

**2. Validar duplicados**

```python
duplicados = tabla.duplicated()
numero_duplicados = duplicados.sum()
print(f"Número de registros duplicados: {numero_duplicados}")
```

**3. Validar registros Nulos**

```python
nulos_por_columna = tabla.isnull().sum()
print(nulos_por_columna)
```

**4. Validar Outliers para el campo Edad**

```python
import numpy as np

# Reemplazar "Sin información" y "Menor de 1 año" por NaN para poder calcular los Outliers 
tabla['Edad'] = tabla['Edad'].replace({'Sin información': np.nan, 'Menor de 1 año': np.nan})

# Convertir la columna a tipo numérico, forzando errores a NaN
tabla['Edad'] = pd.to_numeric(tabla['Edad'], errors='coerce')

media_edad = tabla['Edad'].mean()
tabla['Edad'].fillna(media_edad, inplace=True)
```

```python
# Calcular el IQR
Q1 = tabla['Edad'].quantile(0.25)
Q3 = tabla['Edad'].quantile(0.75)
IQR = Q3 - Q1

# Definir límites para identificar outliers
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

# Filtrar outliers
print("Limite inferior Edad:", limite_inferior)
print("Limite superior Edad:", limite_superior)
outliers = tabla[(tabla['Edad'] < limite_inferior) | (tabla['Edad'] > limite_superior)]

# Eliminar outliers del DataFrame original
tabla_cleaned = tabla[~((tabla['Edad'] < limite_inferior) | (tabla['Edad'] > limite_superior))]
```

```python
print("Número de outliers:", len(outliers))
print("Tabla normal:", len(tabla))
print("Tabla quitando Outliers:", len(tabla_cleaned))
```

**5. Como en el dataset no existen datos Nulos, si existen datos con el dato "Sin informacion" por lo que se busca los registros con mayor cantidad de columnas con este dato para limpiar**

```python
# Contar cuántas columnas tienen el valor "Sin información"
conteo_sin_info = tabla_cleaned.apply(lambda x: (x == "Sin Información").sum(), axis=1)
conteo_sin_info.max()
```

```python
umbral = 4

# Filtrar registros
registros_filtrados = tabla_cleaned[conteo_sin_info > umbral]
print("Registros con mayoría de 'Sin información':")
registros_filtrados
```

```python
# Eliminar registros donde la mayoría de las columnas son 'Sin información'
umbral = 4
tabla_cleaned2 = tabla_cleaned[conteo_sin_info <= umbral]
print("Tabla después de eliminar registros con mayoría de 'Sin información':",len(tabla_cleaned2))
tabla_cleaned2
```
